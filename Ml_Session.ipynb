# Calculation the error (sse)

```python
def error(b,m,points):
    totalError=0
    for i in range(0,len(points)):
        x=points[i,0]
        y=points[i,1]
        totalError+=(y-(m*x+b))**2
    return totalError/float(len(points))
```

# Calculating Gradient

```python
def step_gradient(b_current,m_current,points,learningRate):
    b_gradient=0
    m_gradient=0
    N=float(len(points))
    for i in range(0,len(points)):
        x=points[i,0]
        y=points[i,1]
        b_gradient+=-(2/N)*(y-(m_current*x+b_current))
        m_gradient+=-(2/N)*x*(y-(m_current*x+b_current))
    new_b=b_current-(learningRate*b_gradient)
    new_m=m_current-(learningRate*m_gradient)
    return [new_b,new_m]
```

```python
def gradient(points,starting_b,starting_m,learning_rate,num_iterations):
    b=starting_b
    m=starting_m
    for i in range(num_iterations):
        print("At iteration",i,"Bias is:",b,"Slope is:",m,"Error is:",error(b,m,points))
        [b,m]=step_gradient(b,m,array(points),learning_rate)
    return [b,m]
```

# Fitting a Linear Regression Model

```python
def linearregression():
    try: 
        points=genfromtxt('data.csv',delimiter=",")
        #hyperparameters
        learning_rate=0.0001
        initial_b=0
        initial_m=0
        num_iterations=1000
        [values.final_b,values.final_m]=gradient(points,initial_b,initial_m,learning_rate,num_iterations)
    except:
        print(e)
```
